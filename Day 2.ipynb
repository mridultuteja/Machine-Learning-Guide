## Linear Regression with one variable
Simple linear regression can be used to model a linear relationship between one response variable and one explanatory variable. It is also called as univariate linear regression.
A continuous output variable is a real-value, such as an integer or floating point value. These are often quantities, such as amounts and sizes.

If we think of it in one variable all we are trying to do is fit a line through scattered points on an x-y plane by changing parameters on x
in equation 

`θ0x +  θ1 = y`

So we create a hypothesis function to fit throughout our dataset

hθ(x) = θ0 + θ1x  
Our main aim is here to choose θ0 and θ1 such that hθ(x) is close to y for all of our training samples  
or we need to find the smallest difference between `hθ(x) and y`

### Cost function  
It can be defined as the root mean square difference between the actual value and the calculated value of parameters of hθ(x)  

`J(θ0,θ1) = 1/2m*(hθ(x)-y)^2`  
So in order to get best values to fit our training example value opf our cost function should be minimum for values of θ0 and θ1.  
#### Simplest case
we assume that `θ0 = 0` and `θ1 =1`  
we get a line

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x_0 = np.array([[1],[2],[3]])
y_0 = np.array([[1],[2],[3]])
plt.figure()
plt.plot(x_0,y_0,'x')
plt.plot(x_0,y_0)
plt.grid(True)
plt.show()
```

here J(θ1) = 0

#### Now  
we assume that `θ0 = 0` and `θ1 =0.5`  
we get a line

```python
x_1 = np.array([[1],[2],[3]])
y_1 = np.array([[0.5],[1],[1.5]])
plt.figure()
plt.plot(x_0,y_0,'x')
plt.plot(x_1,y_1)
plt.grid(True)
plt.show()
```

here J(θ1) = 0.58

```python
import sklearn
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
X = np.array([[6], [8], [10], [14], [18],[24]])
y =np.array([[7], [9], [13], [17.5], [18],[23]])
df = pd.DataFrame(np.vstack((X.reshape(1,6),y.reshape(1,6))).T, columns=['Weight (in grams)', 'Price in Rs.' ])
print(df)
```

```python
plt.figure()
plt.title('Weight plotted against price')
plt.xlabel('Weight in grams')
plt.ylabel('Price in Rs.')
plt.plot(X, y, 'b.')
plt.axis([0, 25, 0, 25])
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
print ('A 12 gram will cost: Rs.%.2f' % model.predict([[12]])[0])
end_point = [[0],[25]]
prediction = [ model.predict([i])[0] for i in end_point]
plt.title('Price Plotted against weight')
plt.xlabel('Weight in gram')
plt.ylabel('Price in Rs.')
plt.axis([0, 25, 0, 25])
plt.grid(True)
plt.plot(end_point, prediction)
plt.plot(X, y, 'k.')
plt.show()
```

The sklearn.linear_model.LinearRegression class is an estimator. Estimators predict a value based on the observed data. In scikit-learn, all estimators implement the fit() and predict() methods. The former method is used to learn the parameters of a model, and the latter method is used to predict the value of a response variable for an explanatory variable using the learned parameters
